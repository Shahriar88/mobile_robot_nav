```python
import gymnasium as gym
import mobile_robot_nav
import pygame
```

# Example 01

```python
env = gym.make(
    "MobileRobotNav-v0",
    render_mode="human",
    planner_mode="modified_astar",
    lambda1=1.0,
    lambda2=0.8,
    planning_angle_candidates_deg=[-5, 0, 5],
    require_rotation_clearance=True,
    seed=123,
)
```

```python
running = True
next_episode = False

num_episodes = 20
max_steps_per_episode = 2000

for episode in range(num_episodes):

    obs, info = env.reset()

    print(
        f"Episode {episode+1} started | "
        f"planner={env.unwrapped.planner_mode} | "
        f"lambda1={env.unwrapped.lambda1:.2f} | "
        f"lambda2={env.unwrapped.lambda2:.2f}"
    )

    for step in range(max_steps_per_episode):

        next_episode = False

        for event in pygame.event.get():
            if event.type == pygame.QUIT:
                running = False

            elif event.type == pygame.KEYDOWN:

                if event.key == pygame.K_q:
                    running = False

                elif event.key == pygame.K_n:
                    next_episode = True

                # optional: switch planner live for next episode
                elif event.key == pygame.K_1:
                    env.unwrapped.planner_mode = "astar"
                    print("Planner switched to astar for next reset")

                elif event.key == pygame.K_2:
                    env.unwrapped.planner_mode = "modified_astar"
                    print("Planner switched to modified_astar for next reset")

        if not running:
            break

        if next_episode:
            print("Skipping to next episode...")
            break

        action = env.unwrapped.astar_action()
        obs, reward, terminated, truncated, info = env.step(action)

        if terminated or truncated:
            break

    print(
        f"Episode {episode+1} finished | "
        f"steps={step+1} | "
        f"planner={env.unwrapped.planner_mode} | "
        f"path_len={info.get('path_len', 0)} | "
        f"goal_dist={info.get('goal_distance', float('nan')):.3f}"
    )

    if not running:
        break

env.close()
```

# Example 2 : Manual obstacles

```python
env = gym.make(
    "MobileRobotNav-v0",
    render_mode="human",
    planner_mode="modified_astar",
    lambda1=1.0,
    lambda2=0.8,
    planning_angle_candidates_deg=[-5, 0, 5],
    require_rotation_clearance=True,
    seed=123,
)
```

```python
# Manually defined rectangular obstacles
# Each obstacle is defined as a tuple:
# (cx, cy, hx, hy)
#
# cx = x-coordinate of rectangle center (meters in world frame)
# cy = y-coordinate of rectangle center
#
# hx = half-width along x-axis
#      total width of obstacle = 2 * hx
#
# hy = half-height along y-axis
#      total height of obstacle = 2 * hy
#
# The actual rectangle spans:
# x ∈ [cx - hx , cx + hx]
# y ∈ [cy - hy , cy + hy]

manual_obstacles = [

    # Obstacle 1
    # Center located at (2.0, 1.0)
    # Width  = 2 * 0.4 = 0.8
    # Height = 2 * 0.3 = 0.6
    #
    # Rectangle bounds:
    # x ∈ [1.6 , 2.4]
    # y ∈ [0.7 , 1.3]
    (2.0, 1.0, 0.4, 0.3),

    # Obstacle 2
    # Center located at (1.5, 2.5)
    # Width  = 2 * 0.5 = 1.0
    # Height = 2 * 0.2 = 0.4
    #
    # Rectangle bounds:
    # x ∈ [1.0 , 2.0]
    # y ∈ [2.3 , 2.7]
    (1.5, 2.5, 0.5, 0.2),

    # Obstacle 3
    # Center located at (3.0, 3.0)
    # Width  = 2 * 0.3 = 0.6
    # Height = 2 * 0.8 = 1.6
    #
    # Rectangle bounds:
    # x ∈ [2.7 , 3.3]
    # y ∈ [2.2 , 3.8]
    (3.0, 3.0, 0.3, 0.8),
]
```

```python
running = True
next_episode = False

num_episodes = 20
max_steps_per_episode = 2000

for episode in range(num_episodes):

    obs, info = env.reset(options={"obstacles": manual_obstacles})

    print(
        f"Episode {episode+1} started | "
        f"planner={env.unwrapped.planner_mode} | "
        f"lambda1={env.unwrapped.lambda1:.2f} | "
        f"lambda2={env.unwrapped.lambda2:.2f}"
    )

    for step in range(max_steps_per_episode):

        next_episode = False

        for event in pygame.event.get():
            if event.type == pygame.QUIT:
                running = False

            elif event.type == pygame.KEYDOWN:

                if event.key == pygame.K_q:
                    running = False

                elif event.key == pygame.K_n:
                    next_episode = True

                # optional: switch planner live for next episode
                elif event.key == pygame.K_1:
                    env.unwrapped.planner_mode = "astar"
                    print("Planner switched to astar for next reset")

                elif event.key == pygame.K_2:
                    env.unwrapped.planner_mode = "modified_astar"
                    print("Planner switched to modified_astar for next reset")

        if not running:
            break

        if next_episode:
            print("Skipping to next episode...")
            break

        action = env.unwrapped.astar_action()
        obs, reward, terminated, truncated, info = env.step(action)

        if terminated or truncated:
            break

    print(
        f"Episode {episode+1} finished | "
        f"steps={step+1} | "
        f"planner={env.unwrapped.planner_mode} | "
        f"path_len={info.get('path_len', 0)} | "
        f"goal_dist={info.get('goal_distance', float('nan')):.3f}"
    )

    if not running:
        break

env.close()
```
